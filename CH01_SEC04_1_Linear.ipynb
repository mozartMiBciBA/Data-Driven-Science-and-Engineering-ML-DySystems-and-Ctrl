```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [8, 8]
plt.rcParams.update({'font.size': 18})

x = 3 # True slope
a = np.arange(-2,2,0.25)
a = a.reshape(-1, 1)
#print("a:", a, "a.reshape:", a.shape)
b = x*a + np.random.randn(*a.shape) # Add noise
plt.plot(a, x*a, Color='k', LineWidth=2, label='True line') # True relationship ax = b
plt.plot(a, b, 'x', Color='r', MarkerSize = 8, label='Noisy data') # Noisy measurements

U, S, VT = np.linalg.svd(a,full_matrices=False)
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b # Least-square fit
#print()
#print("xtilde")
#print(xtilde)

#####
# Other Two methods of computing regression

# xtilde1 = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
# xtilde2 = np.linalg.pinv(a) @ b
#####

plt.plot(a,xtilde * a,'--',Color='b',LineWidth=4, label='Regression line')

plt.xlabel('a')
plt.ylabel('b')

plt.grid(linestyle='--')
plt.legend()
plt.show()
```
